## About this notebook
This notebook contains an analysis of a particular race with a particular set of competitors. The functions used to run and analyse the simulations are imported from a .py file. The functions can also be found in a separate jupyter notebook where their development is explained.

## Project Aims
This project is a development of two previous projects that simulate motor races. The aim generally is to allow a team to make decisions that will optimize their race result. In the previous projects the question of interest was which is the optimum lap to take a mandatory pitstop. This project models the performance of the car in more detail so that it simulates the movement of a car around a racetrack allowing for the layout of the track, hence the car will brake approaching corners and accelerate out of them. As a result this allows more questions to be investigated.

In addition to investigating the optimum lap for a pitstop, possible questions are:
1. Should the car be set up to maximize top speed or maximize cornering speed for a particular track?
2. How should the car be developed to maximize performance on the upcoming tracks?
3. What is the best balance between preserving the tyres and driving aggressively?

## Model Features

Compared to the previous iterations of Race_Simulation projects the key features of this model are:

1. Track Layout
Tracks are represented by a Class object with individual corners on each track also represented by Class objects.

1. Overtaking - possibilities are:
Speed advantage required to overtake / run side-by-side. Advantage required determined by track location (straight/corner).

For straight, corner.overtake = 0 (ie any speed advantage will allow the car to overtake)
For an open corner, corner.overtake = 2 (ie an advantage of 2m/s will allow car to overtake)
For a tight corner, corner.overtake = 5
For a very tight corner, corner.overtake = 10 (ie a car will only be able to overtake a much slower car)


2. Positions

3. Pit stops

4. Competitor generation function

5. Qualifying function

6. change race_df position and tyre_wear to float

7. Fuel effect

```python
# Next we will determine the optimal lap for our pit stop based on current car set-up
# repeat the simulation for each potential pit_lap

# create timer
t = rsf.Timer()

# set parameters
grid_offset=5

# create a dataframe to store the simulation results
results_df = pd.DataFrame()

# create lists to store simulation results
#race_df_list = []
#car_df_list = []

# create dict to store race_df and car_df
race_car_df_dict = {}

# set pit_lap_list
pit_lap_list = [8, 9, 10, 11]

# iterate through possible pit_laps
for i in pit_lap_list:
    
    # start timer
    t.start()
    
    # iterate through cars
    for pos in start_grid_dict.keys():
        
        # set car
        car = start_grid_dict[pos]
        
        # reset car attributes in preparation for race
        car.speed = 0
        car.distance_travel = 0
        car.tyre_wear = 0
        car.tyre_deg = 0
        car.tyre_perf = 1
        car.tyre_corner_penalty = 0
        car.lap_count = 0
        car.stopped = False
        car.in_pit = False
        car.pit_duration = 0
        car.next_corner = 0
        
        # set car.distance according to grid position
        car.distance = -pos * grid_offset
        
        # set pit lap for team car
        if car.name == 'Python Racing 1':
            
            car.pit_lap = i
            
            # update car_df
            car_df.loc[(car_df.Car == car.name), 'Pit_Lap'] = car.pit_lap
            
        # debug option
        # print('{} pit_lap = {}'.format(car, car.pit_lap))
            
    # run race
    race_car_df_dict[i] = rsf.race(track=apricot_hill, race_laps=20, start_grid_dict=start_grid_dict, car_df=car_df, \
                           car_length_val=5, time_period=1, race_time_limit=False, debug_val=0)
    
    # save team car result to results_df
    results_df = results_df.append(race_car_df_dict[i][1].iloc[0])
    
    # append race_df and car_df to list
    # race_df_list.append(sim_race_df)
    # car_df_list.append(sim_car_df)
    
    # stop timer
    t.stop()
    
    # print statement
    print(f'Simulation for pit lap {i} complete in {t.elapsed_time} seconds.')
    
# return results_df
results_df
```

In most motor racing categories it is possible to change the set-up of the car to change it's performance characteristics, for example higher downforce will improve high speed cornering performance but will reduce maximum speed. For this project I will assume some simple set-up options.

Cornering / Max_Speed: improving cornering performance by 1% will decrease max_speed by 1% and vice-versa.

Max_Accel + Max_Brake / Max_Tyre_Life: improving max_accel or max_brake by 1% will reduce max_tyre_life by 1% and vice-versa.

Also driving style can be changed, more aggressive driving increases cornering speed but also increases tyre wear.

In order to determine how best to set up the car to optimize our race result there are two possible methods. Firstly we could use the available information (previous experience, looking at the cars that are finishing ahead of us, ie domain knowledge) to identify a possible set-up direction and test it with simulations. Alternatively we could run simulations with all possible combinations of the set-up options.

If we assume a one percent adjustment is possible in either direction for each set-up option we have 27 possible set-up combinations, that would take approximately an hour to run on my laptop. If we allowed a one or two percent adjustment in either direction there would be 125 possible combinations with would take approximately 5 hours to run. Allowing for smaller steps or greater overall adjustment would increase the possible combinations further.

In a real world example the possible combinations would likely be in the hundreds or thousands. For that reason and to avoid long run times we will initially try to determine a set-up direction by examining the race results. I will assume a maximum adjustment of 2% is possible to Cornering, Max_Speed, Max_Accel, Max_Brake and Drive_Style. Changing both Max_Accel and Max_Brake in the same direction by 2% would result in a 4% change in Max_Tyre_Wear.

Having found a promising set-up direction we can now try a limited grid search to test various small changes to see if there is a better option. In this situation we are essentially trying to beat Car_14 to finish in second place.

```python
# create a method to perform a grid search

# set progress counter
iter = 0

# create timer
t = rsf.Timer()

# set number of steps either side of the initial parameter value 
step = 1

# set size of step as percentage of default parameter value
step_perc = 0.01

# set default parameter values
cornering_def = 1.0
max_speed_def = 70
max_accel_def = 10.0
max_brake_def = -20
max_tyre_life_def = 75000
drive_style_def = 1.0

# set initial set-up parameter values
cornering_init = 1.01
max_speed_init = 69.3
max_accel_init = 10.0
max_brake_init = -19.8
max_tyre_life_init = 75750
drive_style_init = 1.0

# set list of pit_laps
pit_lap_list = [9, 10]

# create a dataframe to store the results
grid_cols = ['Max_Accel', 'Max_Brake', 'Max_Speed', 'Max_Tyre_Life', 'Cornering', 'Drive_Style', \
             'Pit_Lap', 'Finish_Position']

grid_results = pd.DataFrame(columns = grid_cols)

# iterate through first parameter pair values
for k in range(-step, step + 1):
    
    # set first parameter pair (Max_Accel/Max_Tyre_Life)
    max_accel = max_accel_init + (max_accel_def * (k * step_perc))
    max_tyre_life = max_tyre_life_init - (max_tyre_life_def * (k * step_perc))
    
    # iterate through second parameter pair values
    for m in range(-step, step + 1):
        
        # set second parameter (Drive_Style)
        drive_style = drive_style_init + (drive_style_def * (m * step_perc))
        
        # iterate through pit_lap list
        for n in pit_lap_list:
            
            # set pit_lap
            pit_lap = n
            
            # set other parameters
            cornering = cornering_init
            max_speed = max_speed_init
            max_brake = max_brake_init
            
            # start timer
            t.start()
            
            # create team car
            car_list, car_df = rsf.create_team_car(team_name='Python Racing 1', max_accel=max_accel, max_brake=max_brake, \
                                               max_speed=max_speed, max_tyre_life=max_tyre_life, cornering=cornering, \
                                               drive_style=drive_style, pit_lap=pit_lap, box_time=3)

            # set seed for random numbers
            random.seed(3)
            
            # create competitors
            car_list, car_df = rsf.create_competitors(car_list=car_list, car_df=car_df, no_comp=19, mean_max_accel=10.0, \
                                                  std_max_accel=0.2, mean_max_brake=-20.0, std_max_brake=0.4, \
                                                  mean_max_speed=70.0, std_max_speed=1.0, mean_max_tyre_life=75000, \
                                                  dif_max_tyre_life=3000, mean_cornering=1.0, std_cornering=0.01, \
                                                  mean_drive_style=1.0, std_drive_style=0.01, pit_lap_list=[9,10,11], \
                                                  box_time=3)

            # set starting grid
            start_grid_dict, car_df = rsf.qualifying(car_list=car_list, car_df=car_df, track=apricot_hill, grid_offset=5)
            
            # run race
            race_df, car_df = rsf.race(track=apricot_hill, race_laps=20, start_grid_dict=start_grid_dict, car_df=car_df, \
                                   car_length_val=5, time_period=1, race_time_limit=False, debug_val=0)
            
            # save team car result in grid_results
            grid_results = grid_results.append({'Max_Accel': max_accel, 'Max_Brake': max_brake, 'Max_Speed': max_speed, \
                                                 'Max_Tyre_Life': max_tyre_life, 'Cornering': cornering, \
                                                'Drive_Style': drive_style, 'Pit_Lap': pit_lap, \
                                                'Finish_Position': car_df['Finish_Position'][0]}, \
                                               ignore_index=True)
            
            # stop timer
            t.stop()
            
            # print progress statement
            iter += 1
            print(f'Iteration {iter} of {((1 + (step * 2))**2) * len(pit_lap_list)} complete in {t.elapsed_time} seconds.')
            
# print grid_results
grid_results
            
```
